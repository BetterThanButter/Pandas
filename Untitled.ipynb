```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', sep = ";")
header = list(df.columns.values)
```

```python
df
```

```python
result = []
prep_set = set()
for index1, value1 in enumerate(df.values):
    for index2, value2 in enumerate(value1):
        new_data = []
        data = value2.split('#')
        new_data.append(data[0])
        new_data.append(1)
        new_data.append(1)
        new_data.append(data[1])
        new_data.append(header[index2])
        prep_set.add(data[1])
      #  print(new_data)
        if(new_data[0] != ""):
            result.append(new_data)
```

```python
new_df = pd.DataFrame(data =result)
```

```python
new_df.to_csv('greg_data.csv', sep=";")
```

```python
gg = ['00','01','02','03','04','05','10','11','12','13','14','15','20','21','22','23','24','25','30','31','32','33','34','35','40','41','42','43','44','45']
```

```python
prep_list = list(prep_set)
teachers_list = []
for index, value in enumerate(prep_list):
    new_row = []
    new_row.append(value)
    new_row = new_row + gg
    teachers_list.append(new_row)
```

```python
new_teachers = pd.DataFrame(data = teachers_list)
new_teachers.head(5)
```

```python
new_teachers.to_csv('teachers_final.csv', header=False,sep = ";", index=False, line_terminator=";\n")
```
