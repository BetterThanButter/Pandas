```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', sep = ";")
df.head()
```

```python
df['hash'] = df['Surname'] + " " + df['Name'].apply(lambda x: x[0:1:]) + '.' + df['Middle_name'].apply(str).apply(lambda x: x[0:1:])+'.'
```

```python
df.to_csv('data_hashed.csv', sep = ";", index= False)
```
